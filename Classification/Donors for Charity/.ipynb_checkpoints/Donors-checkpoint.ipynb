# Preparing data for preprocessing

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
numerical = ['age', 'education-num', 'capital-gain', 'capital-loss']
df[numerical]=scaler.fit_transform(df[numerical])
```

### This is done to make more accurate calculations while using KNN for classification

```python
y = df['Income']
```

```python
y.head()
```

```python
df.drop(['Income'], axis=1, inplace=True)
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y = y.apply(lambda x:0 if x=='<=50K' else 1)
```

```python
y
```

```python
df
```

#### One Hot Encoding (makes the categorical variables more expressive)

```python
x = pd.get_dummies(df)
x
```

```python
df['workclass'].unique()
```

```python
list(x.columns)
```

# Training and Testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size=0.3, random_state=0)
```

```python
X_train
```

```python
X_test
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(min_samples_split=90,max_depth=11,criterion='gini')
dt.fit(X_train, Y_train)
dt.score(X_train, Y_train)
```

```python
prediction = dt.predict(X_test)#This is the trained model
```

```python
list(prediction)
```

```python
list(Y_test)
```

## Comparing with the actual data

```python
from sklearn.metrics import accuracy_score
accuracy_dt = accuracy_score(Y_test, prediction)*100
```

```python
accuracy_dt
```

```python
dt.predict(X_test[0:1])
```

# Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters= {'min_samples_split':range(10,500,20), 'max_depth':range(1,20,2), 'criterion':['gini', 'entropy']}
clf_tree = DecisionTreeClassifier()
clf = GridSearchCV(clf_tree,parameters, cv=5)
clf.fit(X_train, Y_train)
```

```python
clf.best_estimator_
```

```python
clf_predict = clf.predict(X_test)
```

#### Observe the increase in accuracy

```python
accuracy_score(clf_predict,Y_test)
```

`min_samples_split=90,max_depth=11,criterion='gini'`
# then we passed these parameters into the decision tree model

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression(C=100, solver='newton-cg')
log.fit(X_train, Y_train)
log
```

```python
log_test = log.predict(X_test)
```

```python
log_test
```

```python
accuracy_score(log_test, Y_test)#This is the testing score
```

```python
log.score(X_train, Y_train)#This is the training score
```

```python
model = LogisticRegression(max_iter=10000)
solvers=['newton-cg','lbfgs','liblinear']
penalty = ['l2']
c_values=[100, 10, 0.1, 0.01, 0.001]
parameters = dict(solver=solvers, penalty=penalty, C=c_values)
```

```python
parameters
```

```python
grid_search_log = GridSearchCV(estimator= model, param_grid=parameters, cv=5)
grid_search_log.fit(X_train, Y_train)
```

```python
grid_search_log.best_estimator_
```

```python
grid_log = grid_search_log.predict(X_test)
```

```python
accuracy_score(grid_log, Y_test)
```
